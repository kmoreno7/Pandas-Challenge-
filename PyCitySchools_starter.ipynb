```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
number_of_schools = len(pd.unique(school_data_complete["school_name"]))
print(number_of_schools)

#Calculate the total number of students
number_of_students = len((school_data_complete["student_name"]))
print(number_of_students)

#Calculate the total budget
individual_school_budget = pd.unique(school_data_complete['budget'])
total_budget = sum(individual_school_budget)
print(total_budget)

#Calculate the average math score
average_math_score = round(school_data_complete['math_score'].mean(),2)
print(average_math_score)

#Calculate the average reading score
average_reading_score = round(school_data_complete['reading_score'].mean(),2)
print(average_reading_score)

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_score = round(len(school_data_complete[school_data_complete["math_score"] >= 70]) / number_of_students * 100, 2)
print(passing_math_score)

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_score = round(len(school_data_complete[school_data_complete["reading_score"] >= 70]) / number_of_students * 100, 2)
print(passing_reading_score)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = round(len(school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]) / number_of_students * 100, 2)
print(overall_passing)

#Create a dataframe to hold the above results

data = {'Summary' : ["number of schools", "number of students", "total budget", "average math score", "average reading score", "% of students with passing math score", "% of students withpassing reading score", "% of students with an overall passing"],
        'Result' : [number_of_schools, number_of_students, total_budget, average_math_score, average_reading_score, passing_math_score, passing_reading_score, overall_passing]}
        

district_summary = pd.DataFrame(data)


print(district_summary)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Pull School Name, School Type, Total students, Total budget from original dataframe 

df = school_data_complete[["school_name","type","size","budget"]]
school_type= df.drop_duplicates(subset = ["school_name"])
school_summary = pd.DataFrame(school_type)
print(school_summary)
```

```python
#Per Student Budget

school_summary["per_student_budget"] = school_summary["budget"]/school_summary["size"]
print(school_summary)
```

```python
#average reading and math scores 

df2 = school_data_complete[["school_name","reading_score","math_score"]]
scores = pd.DataFrame(df2)

school_averages = scores.groupby(['school_name']).mean()

school_summary = pd.merge(school_summary, school_averages, how="left", on=["school_name", "school_name"])

school_summary = school_summary.sort_values(by=["school_name"])

print(school_summary)
```

```python
#Passing math, reading, and both

#Passing Math
df3 = school_data_complete[["school_name","math_score"]]
math_scores = pd.DataFrame(df3)

passing_math_scores = (math_scores[(math_scores["math_score"] >= 70)])

students_passing_math = passing_math_scores.groupby(['school_name']).count()

students_passing_math.rename(columns = {'math_score':'students_passing_math'}, inplace = True)

#Passing Reading
df3 = school_data_complete[["school_name","reading_score"]]
reading_scores = pd.DataFrame(df3)

passing_reading_scores = (reading_scores[(reading_scores["reading_score"] >= 70)])

students_passing_reading = passing_reading_scores.groupby(['school_name']).count()

students_passing_reading.rename(columns = {'reading_score':'students_passing_reading'}, inplace = True)

#Passing Both 

passing_both_scores = (scores[(scores["reading_score"] >= 70) & (scores["math_score"] >= 70)])

students_passing_both = passing_reading_scores.groupby(['school_name']).count()

students_passing_both.rename(columns = {'reading_score':'passing_both'}, inplace = True)

#Creating new dataframe 
school_summary = pd.merge(school_summary, students_passing_math, how="left", on=["school_name", "school_name"])
school_summary = pd.merge(school_summary, students_passing_reading, how="left", on=["school_name", "school_name"])
school_summary = pd.merge(school_summary, students_passing_both, how="left", on=["school_name", "school_name"])

#Adding% passing math, % passing reading , % overall passing

school_summary["percent_passing_math"] = round((school_summary['students_passing_math']/school_summary["size"]) * 100, 2)
school_summary["percent_passing_reading"] = round((school_summary['students_passing_reading']/school_summary["size"]) * 100, 2)
school_summary["percent_passing_both"] = round((school_summary["passing_both"]/school_summary["size"]) * 100, 2)

print(school_summary)

```

```python
#creating date frame for the next two questions

df4 = school_summary[["school_name","percent_passing_both"]]
schools_passing = pd.DataFrame(df4)

print(schools_passing)
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_preforming_schools = schools_passing.sort_values('percent_passing_both', ascending=False).head(5)
print(top_preforming_schools)
```
